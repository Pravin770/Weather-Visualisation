# Importing libraries

```python
import pandas as pd 
```

# Function to Read File

```python
def read_file(fileName):
    
    with open(fileName, 'r', encoding="utf-8") as file:
        a=file.readlines() 
         
        a=([sub.split(",") for sub in a])
        kols = [i.replace('\n','').replace('"',"") for i in a[0]]
        datalist = [i for i in a[1:]] 
         
        return kols,datalist
        
```

# Reading Files

```python
#reading files
#Change the File directory path according to the location.
kols,datalist=read_file(r'C:\Users\The Best\Downloads\lab1-data\barometer-last-year.csv')
Baro_df = pd.DataFrame(data=datalist,columns=kols )
Baro_df = Baro_df.replace('\n','', regex=True)
Baro_df = Baro_df.replace('"','', regex=True)
Baro_df['Baro'] = Baro_df['Baro'].astype(float, errors = 'raise')
Baro_Min=Baro_df['Baro'].min()
Baro_Max=Baro_df['Baro'].max()
Baro_Mean=Baro_df['Baro'].mean()
Baro_std=Baro_df['Baro'].std()
Baro_df=Baro_df.rename(columns={Baro_df.columns[0]: 'DT'})
Baro_df['DT'] = pd.to_datetime(Baro_df['DT'], format="%Y-%m-%d %H:%M:%S")
Baro_df
```

```python
#Change the File directory path according to the location.

kols,datalist=read_file(r'C:\Users\The Best\Downloads\lab1-data\indoor-temperature-last-year.csv')
Ind_temp_df = pd.DataFrame(data=datalist,columns=kols)
Ind_temp_df = Ind_temp_df.replace('"','', regex=True)
Ind_temp_df['Humidity'] = Ind_temp_df['Humidity'].astype(int, errors = 'raise')
Ind_temp_df['Temperature'] = Ind_temp_df['Temperature'].astype(float, errors = 'raise')
Ind_temp_df['Temperature']=round(Ind_temp_df['Temperature'], 2)
Ind_temp_df['Temperature_range (low)'] = Ind_temp_df['Temperature_range (low)'].astype(float, errors = 'raise')
Ind_temp_df['Temperature_range (high)'] = Ind_temp_df['Temperature_range (high)'].astype(float, errors = 'raise')
Ind_temp_df=Ind_temp_df.rename(columns={Ind_temp_df.columns[0]: 'DT'})
Ind_temp_df['DT'] = pd.to_datetime(Ind_temp_df['DT'], format="%Y-%m-%d %H:%M:%S")
Ind_temp_df=Ind_temp_df.rename(columns={"Temperature": "Indoor Temperature", "Temperature_range (low)": "Indoor Temperature_range (low)","Temperature_range (high)": "Indoor Temperature_range (high)"})
 

Ind_temp_df
```

```python
#Change the File directory path according to the location.

kols,datalist=read_file(r'C:\Users\The Best\Downloads\lab1-data\outside-temperature-last-year.csv')
Out_temp_df = pd.DataFrame(data=datalist,columns=kols)
Out_temp_df = Out_temp_df.replace('"','', regex=True)
Out_temp_df['Temperature'] = Out_temp_df['Temperature'].astype(float, errors = 'raise')
Out_temp_df['Temperature']=round(Out_temp_df['Temperature'], 2)
Out_temp_df['Temperature_range (low)'] = Out_temp_df['Temperature_range (low)'].astype(float, errors = 'raise')
Out_temp_df['Temperature_range (high)'] = Out_temp_df['Temperature_range (high)'].astype(float, errors = 'raise')
Out_temp_df=Out_temp_df.rename(columns={Out_temp_df.columns[0]: 'DT'})
Out_temp_df['DT'] = pd.to_datetime(Out_temp_df['DT'], format="%Y-%m-%d %H:%M:%S")
Out_temp_df=Out_temp_df.rename(columns={"Temperature": "Outdoor Temperature", "Temperature_range (low)": "Outdoor Temperature_range (low)","Temperature_range (high)": "Outdoor Temperature_range (high)"})
Out_temp_df
```

```python
#Change the File directory path according to the location.

kols,datalist=read_file(r'C:\Users\The Best\Downloads\lab1-data\rainfall-last-year.csv')
Rainfall_temp_df = pd.DataFrame(data=datalist,columns=kols)
Rainfall_temp_df = Rainfall_temp_df.replace('"','', regex=True)
Rainfall_temp_df['mm'] = Rainfall_temp_df['mm'].astype(float, errors = 'raise')
Rainfall_temp_df=Rainfall_temp_df.rename(columns={Rainfall_temp_df.columns[0]: 'DT'})
Rainfall_temp_df['DT'] = pd.to_datetime(Rainfall_temp_df['DT'], format="%Y-%m-%d %H:%M:%S")
Rainfall_temp_df=Rainfall_temp_df.rename(columns={"mm": "Rainfall_mm"})
Rainfall_temp_df
```
